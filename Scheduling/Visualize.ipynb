```python
from matplotlib import pyplot as plt
```

```python
results = []
with open("r_GA.txt") as f:
    for i in f:
        violation, time = i.split()
        results.append([int(violation), float(time)])
```

```python
results2 = []
with open("r_SHC.txt") as f:
    for i in f:
        violation, time = i.split()
        results2.append([int(violation), float(time)])
```

```python
x1 = [i[0] for i in results]
x2 = [i[0] for i in results2]
plt.figure(dpi=250)
plt.plot(list(range(10, 10 + len(x1)*10, 10)), x1)
plt.plot(list(range(10, 10 + len(x1)*10, 10)), x2)
plt.xlabel("the number of tasks")
plt.ylabel("the number of violations")
plt.legend(["GA", "SHC"])
```

```python
x1 = [i[1] for i in results]
x2 = [i[1] for i in results2]
plt.figure(dpi=250)
plt.plot(list(range(10, 10 + len(x1)*10, 10)), x1)
plt.plot(list(range(10, 10 + len(x1)*10, 10)), x2)
plt.xlabel("the number of tasks")
plt.ylabel("Running times(s)")
plt.legend(["GA", "SHC"])
```
